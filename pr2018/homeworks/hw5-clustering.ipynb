## 数据集 

```python
iris = datasets.load_iris()
points = iris.data[:,:2]
fig, ax = plt.subplots()
ax.plot(points[iris.target == 0,0], points[iris.target == 0,1], 'ro', 
        points[iris.target == 1,0], points[iris.target == 1,1], 'bx',
        points[iris.target == 2,0], points[iris.target == 2,1], 'k*')
ax.set_xlabel('feature 1')
ax.set_ylabel('feature 2')
ax.set_title('iris dataset (%d examples)' % len(points))
```
