1. 用skimage读取图像文件
2. 用matplotlib显示图像

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import NoNorm
from skimage.io import imread
%matplotlib inline
```

```python
# source: https://upload.wikimedia.org/wikipedia/commons/thumb/f/fa/Weeki_Wachee_spring_10079u.jpg/480px-Weeki_Wachee_spring_10079u.jpg
imfname = 'data/480px-Weeki_Wachee_spring_10079u.jpg'
img = imread(imfname)
pixels = img.flatten()

gammas = [1, 0.25, 0.5, 2]
fig, axes = plt.subplots(nrows=1, ncols=len(gammas), figsize=(16,8))

for i in range(len(gammas)):
    enhanced = map(lambda x:int(((x/255.0)**gammas[i])*255), pixels)
    enhanced = np.reshape(enhanced, img.shape)
    axes[i].imshow(enhanced, cmap='gray', norm=NoNorm())
    axes[i].set_title(r'$\gamma=%g$'% gammas[i], fontsize=20)
    axes[i].set_axis_off()
```
